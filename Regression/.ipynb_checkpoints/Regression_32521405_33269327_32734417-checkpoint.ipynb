```python
from pandas import read_excel
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.api import Holt,ExponentialSmoothing
from statsmodels.formula.api import ols
import numpy as np
```

```python
Collateddf = read_excel('MSTA.xlsx', sheet_name='regression_model', header=0, 
                     squeeze=True)

#Creating test (12 months) and train (288 months) for running the regression model

Collateddf_train=Collateddf[0:288]
Collateddf_test=Collateddf[289:300]
```

#### Creating scatter plot to check for correlation

```python
pd.plotting.scatter_matrix(Collateddf[['MSTA','GMAF','ET12','CH4']])
plt.show()
```

####Creating correlation matrix to check for collinearity

```python
CorrelationMatrix = Collateddf[['MSTA','GMAF','ET12','CH4']].corr()
CorrelationMatrix
```

#### Assigning train variables for the regression model

```python
MSTA = Collateddf_train.MSTA
GMAF = Collateddf_train.GMAF
ET12 = Collateddf_train.ET12
CH4 = Collateddf_train.CH4
D1=Collateddf_train.D1
D2=Collateddf_train.D2
D3=Collateddf_train.D3
D4=Collateddf_train.D4
D5=Collateddf_train.D5
D6=Collateddf_train.D6
D7=Collateddf_train.D7
D8=Collateddf_train.D8
D9=Collateddf_train.D9
D10=Collateddf_train.D10
D11=Collateddf_train.D11
Time=Collateddf_train.Time2

```

#### Generating CH4, GMAF and ET12 forecast using the best exponential smoothing method for the respective time series

```python
HoltMethod1 = ExponentialSmoothing(CH4, seasonal_periods=12, trend='add', seasonal='add').fit()
Holtcast1=HoltMethod1.forecast(48)
HoltMethod1.fittedvalues.plot(color='blue')
```

```python
HoltMethod2 = ExponentialSmoothing(np.sqrt(GMAF), seasonal_periods=12, trend='add', seasonal='add').fit()
Holtcast2=np.square(HoltMethod2.forecast(48))
HoltMethod2.fittedvalues.plot(color='blue')
```

```python
HoltMethod3 = ExponentialSmoothing(ET12, seasonal_periods=12, trend='add', seasonal='mul').fit()
Holtcast3=HoltMethod3.forecast(48)
HoltMethod3.fittedvalues.plot(color='blue')
```

## Regression model 1: MSTA = 0 + 1(CH4) + 2(GMAF)  + 3(ET1.2) + 

```python
formula = 'MSTA ~ CH4+GMAF+ET12'

# ols generate statistics and the parameters b0, b1, etc., of the model
results = ols(formula, data=Collateddf_train).fit()
results.summary()
```

## Regression model 2: MSTA = 0 + 1(CH4)+ 2(GMAF)  + 3(ET1.2)+ 4(D1)+ 5(D2)+ 6(D3)+ 7(D4)+ 8(D5)+ 9(D6)+ 10(D7)+ 11(D8)+ 12(D9)+ 13(D10)+ 14(D11)+ 

```python
formula = 'MSTA ~ CH4+GMAF+ET12+D1+D2+D3+D4+D5+D6+D7+D8+D9+D10+D11'

# ols generate statistics and the parameters b0, b1, etc., of the model
results = ols(formula, data=Collateddf_train).fit()
results.summary()
```

## Regression model 3: MSTA =0 + 1(CH4)+ 2(GMAF)  + 3(ET1.2)+ 4(D1)+ 5(D2)+ 6(D3)+ 7(D4)+ 8(D5)+ 9(D6)+ 10(D7)+ 11(D8)+ 12(D9)+ 13(D10)+ 14(D11)+ 15(t)+ 

```python
formula = 'MSTA ~ CH4+GMAF+ET12+D1+D2+D3+D4+D5+D6+D7+D8+D9+D10+D11+Time2'

# ols generate statistics and the parameters b0, b1, etc., of the model
results = ols(formula, data=Collateddf_train).fit()
results.summary()
```

## Model 4: MSTA = 0 + 1(CH4) + 2(D2) + 3(t)+ e   

```python
formula = 'MSTA ~ CH4+D2+Time2'

# ols generate statistics and the parameters b0, b1, etc., of the model
results = ols(formula, data=Collateddf_train).fit()
results.summary()
```

## Analysing key metrics we have decided to do the forecast only for model 4


### We use the outputs obtained from HoltMethod1, HoltMethod2 and HoltMethod3 which predicts values for CH4, GMAF and ET12 respectively

### The sheet regression_pred in the file MSTA_32521405_33269327_32734417 has been updated with all the available actual values till 2021 and for values that are not available has been filled with exponential smoothing methods for CH4, GMAF and ET12.

```python
Collateddf_forecast = read_excel('MSTA_32521405_33269327_32734417.xlsx', sheet_name='regression_pred', header=0, 
                     squeeze=True)

a1 = np.array(Collateddf_forecast.CH4[:324])
D2=Collateddf_forecast.D2[:324]
Time=Collateddf_forecast.Time2[:324]

#a2 = np.array(Collateddf_forecast.GMAF[:324])
#a3 = np.array(Collateddf_forecast.ET12[:324])
#D1=Collateddf_forecast.D1[:324]
#D3=Collateddf_forecast.D3[:324]
#D4=Collateddf_forecast.D4[:324]
#D5=Collateddf_forecast.D5[:324]
#D6=Collateddf_forecast.D6[:324]
#D7=Collateddf_forecast.D7[:324]
#D8=Collateddf_forecast.D8[:324]
#D9=Collateddf_forecast.D9[:324]
#D10=Collateddf_forecast.D10[:324]
#D11=Collateddf_forecast.D11[:324]
```

#### Getting the intercepts for the regression model 4

```python
b0 = results.params.Intercept
b1 = results.params.CH4
dval2=results.params.D2
timeval=results.params.Time2


#b2 = results.params.GMAF
#b3 = results.params.ET12
#dval1=results.params.D1
#dval3=results.params.D3
#dval4=results.params.D4
#dval5=results.params.D5
#dval6=results.params.D6
#dval7=results.params.D7
#dval8=results.params.D8
#dval9=results.params.D9
#dval10=results.params.D10
#dval11=results.params.D11
```

#### Using the model 4 regression model to forecast MSTA values for the existing time range (Till 2021)

```python
F=a1
for i in range(324):
    F[i] = b0  + a1[i]*b1+ D2[i]*dval2 + timeval*Time[i]
```

#### Using the forecasted regression model to predict the values for 2022

```python
v1=Collateddf_forecast.CH4[324:]
D2_pred=Collateddf_forecast.D2[324:]
time_pred=Collateddf_forecast.Time2[324:]

#v2=Collateddf_forecast.GMAF[324:]
#v3=Collateddf_forecast.ET12[324:]
#D1_pred=Collateddf_forecast.D1[324:]
#D3_pred=Collateddf_forecast.D3[324:]
#D4_pred=Collateddf_forecast.D4[324:]
#D5_pred=Collateddf_forecast.D5[324:]
#D6_pred=Collateddf_forecast.D6[324:]
#D7_pred=Collateddf_forecast.D7[324:]
#D8_pred=Collateddf_forecast.D8[324:]
#D9_pred=Collateddf_forecast.D9[324:]
#D10_pred=Collateddf_forecast.D10[324:]
#D11_pred=Collateddf_forecast.D11[324:]
```

```python
v1=v1.values
D2_pred=D2_pred.values
time_pred=time_pred.values


#v2=v2.values
#v3=v3.values
#D1_pred=D1_pred.values
#D3_pred=D3_pred.values
#D4_pred=D4_pred.values
#D5_pred=D5_pred.values
#D6_pred=D6_pred.values
#D7_pred=D7_pred.values
#D8_pred=D8_pred.values
#D9_pred=D9_pred.values
#D10_pred=D10_pred.values
#D11_pred=D11_pred.values
```

```python
#predicted values
E=v1
for i in range(12):
    E[i] = b0  + v1[i]*b1+ D2_pred[i]*dval2 + timeval*time_pred[i]
    
    
```

#### appending the full range of values of the forecasted MSTA and finding the error by comparing it with the actual values till 2019.

```python
K=np.append(F, E)
MSTA0 = Collateddf_forecast.MSTA
values=MSTA0[0:288] #You can change here also
Error = values - F[0:288]
MSE=sum(Error**2)*1.0/len(F)
MSE
```

#### Creating lowerbound and upperbound values

```python
LowerE = E - 1.282*np.sqrt(MSE)
UpperE = E + 1.282*np.sqrt(MSE)
LowerEfinal=[]
UpperEfinal=[]
LowerEfinal=np.append(LowerEfinal, np.repeat(np.nan, 324))
UpperEfinal=np.append(UpperEfinal, np.repeat(np.nan, 324))
LowerEfinal=np.append(LowerEfinal,LowerE)
UpperEfinal=np.append(UpperEfinal,UpperE)
```

#### plotting the forecasted values against original dataq

```python
from matplotlib.legend_handler import HandlerLine2D

line1, = plt.plot(K, color='red', label='Forecast values')
line2, = plt.plot(MSTA0, color='black', label='Original data')
line3, = plt.plot(LowerEfinal, color='blue', label='Lower forecast')
line4, = plt.plot(UpperEfinal, color='orange', label='Upper forecast')
plt.rcParams["figure.figsize"] = (20,15)
plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})
plt.title('MSTA regression forecast with confidence interval')
plt.show()
```
