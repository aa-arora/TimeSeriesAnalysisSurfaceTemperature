## Importing the libraries used

```python
# pdarima is installed to implement auto_arima.
!pip install pmdarima
from pandas import read_excel
import pandas as pd
import numpy as np
from matplotlib import pyplot
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
import statsmodels.api as sm
from pmdarima import auto_arima
```

```python
# Reading the excel file
df = read_excel('MSTA.xlsx', sheet_name='MSTA', header=0, parse_dates=True, squeeze=True, index_col='Time') 
```

```python
df.head()
```

## Data prepartation
The data is considered from January of 1995 to December 2021. This is the same data considered in the exponential smoothing model.

```python
filtered_df = df.loc[(df.index >= '1995-01-01')]
```

```python
filtered_df.head()
```

#### Dropping columns that are not used
The columns that are not used are - "Lower confidence limit (2.5%)" and "Upper confidence limit (97.5%)". These columns are dropped and the dataframe is updated

```python
filtered_df = filtered_df.drop(["Lower confidence limit (2.5%)", "Upper confidence limit (97.5%)"], axis=1)
```

```python
# The column is renamed to 'Temp' as this is easier to read and understand
filtered_df.rename({'Anomaly (deg C)': 'Temp'}, axis=1, inplace=True)
```

```python
filtered_df
```

### Preliminary Analysis

```python
#Time plot of the original MSTA from 1995
filtered_df.plot(figsize=(15,10))
```

#### From the plot we can see that the data is not stationary. This is further proved by plotting the acf and pacf plots

```python
plot_acf(filtered_df["Temp"], title='ACF of temperature time series', lags=50)
# PACF plot on 50 time lags
plot_pacf(filtered_df["Temp"], title='PACF of temperature time series', lags=50)
pyplot.show()
```

### The stationary charecteristic of the data can also be tested with the help of the adfuller test.
The adfuller test has the null hypothesis <br>H0: The data is not stationary <br> 
The significance level considered is 0.05 (5%)

```python
def adfuller_test(Temp):
    result=adfuller(Temp)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
    
```

```python
adfuller_test(filtered_df['Temp'])
```

## Performing First differencing

```python
filtered_df['First Difference'] = filtered_df['Temp'] - filtered_df['Temp'].shift(1)
```

```python
filtered_df = filtered_df.dropna()
```

```python
# Time plot of the first difference data
filtered_df['First Difference'].plot(figsize=(15,10))
```

We can see from the above time plot that the first difference data is now stationary. <br> This is also proved from the pacf and acf plots below

```python
plot_acf(filtered_df["First Difference"], title='ACF of first difference time series', lags=50)
# PACF plot on 50 time lags
plot_pacf(filtered_df["First Difference"], title='PACF of first difference time series', lags=50)
pyplot.show()
```

```python
# The adfuller test is again performed on the first difference data and we can see that the p-value 
# is lesser than the significance level and thus the null hypothesis is rejected and the data is stationary
adfuller_test(filtered_df['First Difference'])
```

### Model Selection
First, select the ARIMA model by analyzing the acf and pacf plots. We select the ARIMA(0,0,1) model. The model is then analysed with the AIC values

```python
model=sm.tsa.statespace.SARIMAX(filtered_df['Temp'],order=(0, 1, 1), seasonal_order=(0,0,0,0))
results=model.fit()
results.summary()
```

The AIC Value got from the model ARIMA(0,1,1) is -556.337

### AUTOARIMA 
Next we use AUTOARIMA to generate the best ARIMA model. <br> From the results below, it can be concluded that the best model is the ARIMA(2,1,2)(0,0,0)0 model which gives a AIC value of -566.910

```python
stepwise_fit = auto_arima(filtered_df['Temp'], trace=True,
suppress_warnings=True, max_p = 3, max_q = 3)
```

```python
# Fitting the ARIMA model witht the parameter values got from the AUTOARIMA function
model=sm.tsa.statespace.SARIMAX(filtered_df['Temp'],order=(2, 1, 2), seasonal_order=(0,0,0,0))
results=model.fit()
results.summary()
```

### Plotting the forecast using the optimal ARIMA model

```python
start=1
pred=results.get_prediction(start=start, dynamic=False)
pred_ci = pred.conf_int()
ax = filtered_df['1995-01-01':]['Temp'].plot(label='Original data')
pred.predicted_mean.plot(ax=ax, label='ARIMA Forecast', alpha=.7, figsize=(15, 12)) 
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.05)
plt.legend()
plt.show()
```

### Calculating the MSE of the forecast from the ARIMA model plotted above

```python
y_forecasted = pred.predicted_mean
y_truth = filtered_df['2019-01-01':]['Temp']
# Compute the mean square error
MSE = ((y_forecasted - y_truth) ** 2).mean()
print('MSE of the forecasts is {}'.format(round(MSE, 5)))
```

```python
# plotting diagnostics of the model
results.plot_diagnostics(figsize=(15, 12))
```

### Forecasting MSTA values for 2022

```python
# Get forecast 12 steps ahead in future
pred_uc = results.get_forecast(steps=12)
# Get confidence intervals of forecasts
pred_ci = pred_uc.conf_int()
# plotting forecasts ahead
ax = filtered_df['2022-01-01':]['Temp'].plot(label='Original data')
pred_uc.predicted_mean.plot(ax=ax, label='Forecast values', title='Forecast plot with confidence interval', figsize=(15, 12))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.02)
plt.axis(['2022-01-01', '2022-12-01', 0.75, 0.85])
plt.yticks(np.arange(0.77, 0.83, 0.005))
plt.legend()
plt.show()
```
